```python
import numpy as np
import matplotlib.pyplot as plt
import torch
import torchvision
import torchvision.transforms as T
import matplotlib.pyplot as plt
```

```python
cdata = torchvision.datasets.CIFAR10(root='cifar10', download=True)
```

```python
print(cdata)
```

```python
cdata.data.shape
```

```python
cdata.classes
```

```python
len(cdata.targets)
```

```python
fig, axes = plt.subplots(5,5, figsize=(8,6))
for ax in axes.flatten():
  idx = np.random.choice(len(cdata.targets))
  image = cdata.data[idx,:,:,:]
  label = cdata.classes[cdata.targets[idx]]
  ax.imshow(image)
  ax.text(16,0,label, ha='center', fontweight='bold', color='k', backgroundcolor='y')
  ax.axis('off')

plt.tight_layout()
```

```python
## Applying Transformation

Ts = T.Compose([
                T.ToTensor(),
                T.Resize(32*4),
                T.Grayscale(num_output_channels=1)
])

cdata.transform = Ts ##alternatively cdata['transform'] =Ts

print(cdata.data[123,:,:,:].shape)

## adding transform does not change size of original data
```
